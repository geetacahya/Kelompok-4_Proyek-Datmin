# **Penentuan Jumlah Klaster**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

## **Elbow Method**

```python
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt

n_cluster_list = np.arange(2, 11, 1) #daftar eksperimen jumlah cluster
clustering_models = [KMeans(n_clusters=n_cluster).fit(data_min_max_scaled) for n_cluster in n_cluster_list] #daftar model KMeans yang telah dilatih pada data
centroids_list = [cluster_model.cluster_centers_ for cluster_model in clustering_models] #daftar centroid tiap model

distances_to_centroid = [cdist(data_min_max_scaled, centroid, 'euclidean') for centroid in centroids_list]
distance_list = [np.min(D, axis=1) for D in distances_to_centroid]
avgWithinSS = [sum(d) / data_min_max_scaled.shape[0] for d in distance_list]
```

```python
# Membuat plot Elbow Method
plt.figure(figsize=(10, 6))
sns.lineplot(x=n_cluster_list, y=avgWithinSS, marker='o', color='g')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Average Within-Cluster Sum of Squares')
plt.title('Elbow Method for KMeans')
plt.grid(True)
plt.show()
```

```python
# Menginisialisasi model KMeans dengan jumlah cluster yang diinginkan (3)
best_kmeans = KMeans(n_clusters=3)
best_kmeans.fit(data_min_max_scaled)
```

```python
# Menampilkan DataFrame
data_min_max_scaled['cluster'] = best_kmeans.labels_
data_min_max_scaled
```

## **Silhouette Score**

```python
# Daftar eksperimen jumlah cluster
n_cluster_list = np.arange(2, 11, 1)
silhouette_score_list = []

# Melakukan eksperimen untuk setiap jumlah cluster
for cluster in n_cluster_list:
  kmeans = KMeans(n_clusters= cluster)
  kmeans.fit(data_min_max_scaled)
  silhouette_score_list.append(silhouette_score(X= data_min_max_scaled, labels= kmeans.labels_))
```

```python
n_cluster_experiments = pd.DataFrame({
    'Number of Cluster': n_cluster_list,
    'Silhouette Score': silhouette_score_list
})
n_cluster_experiments = n_cluster_experiments.sort_values(by= 'Silhouette Score', ascending= False)
n_cluster_experiments
```

```python
# Membuat plot Silhouette Score
plt.figure(figsize= (10, 6))
sns.lineplot(x=n_cluster_list, y= silhouette_score_list, marker='o', color='r')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for KMeans')
plt.grid(True)
plt.show()
```

# **Principal Component Analysis (PCA)**

```python
from sklearn.decomposition import PCA
import pandas as pd

# Inisialisasi PCA
pca = PCA(n_components=2)  # Mengurangi ke 2 dimensi (Anda bisa mengubah jumlah komponen sesuai kebutuhan)

# Terapkan PCA
data_pca = pca.fit_transform(data_min_max_scaled)

# Mengonversi hasil PCA kembali ke DataFrame
data_pca_df = pd.DataFrame(data_pca, columns=['Principal Component 1', 'Principal Component 2'])

# Menampilkan hasil PCA
display(data_pca_df.head().style.set_caption("Data Setelah PCA"))
```

## Visualisasi Hasil Clustering dengan PCA

```python
import matplotlib.pyplot as plt

# Membuat plot scatter untuk dua komponen utama
plt.figure(figsize=(10, 6))
plt.scatter(data_pca_df['Principal Component 1'], data_pca_df['Principal Component 2'], alpha=0.7)

# Menambahkan label dan judul
plt.title('Visualisasi Hasil Clustering dengan PCA', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=14)
plt.ylabel('Principal Component 2', fontsize=14)
plt.grid()

# Menampilkan plot
plt.show()
```

# **K-Means Clustering**

```python
# Menginisialisasi model KMeans dengan jumlah cluster yang diinginkan (misalnya 3)
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Melatih model KMeans pada data PCA
kmeans.fit(data_pca_df)

# Menambahkan label cluster ke DataFrame PCA
data_pca_df['Cluster'] = kmeans.labels_

# Menampilkan DataFrame dengan label cluster
print(data_pca_df.head())
```

## **Visualisasi Hasil Clustering KMeans pada Data PCA**

```python
# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
plt.scatter(data_pca_df['Principal Component 1'], data_pca_df['Principal Component 2'],
            c=data_pca_df['Cluster'], cmap='viridis', alpha=0.7)

plt.title('Hasil Clustering KMeans pada Data PCA', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=14)
plt.ylabel('Principal Component 2', fontsize=14)
plt.colorbar(label='Cluster')
plt.grid()
plt.show()
```
